### Section 1: A Practical Example

This section explain what is data science, what are the examples based on a data and I'll take a data from Sololearn for this section.

```python
# Basic Setup
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import sklearn
sns.set()
from sklearn.cluster import KMeans
from sklearn import preprocessing
plt.style.use('ggplot')
```

```python
# Let's take the data from Sololearn
us_president = pd.read_csv('https://sololearn.com/uploads/files/president_heights_party.csv', index_col='name')
```

```python
plt.scatter(us_president['age'], us_president['height'])
plt.xlabel('age')
plt.ylabel('height')
plt.title('US President Scatter Plot')
plt.savefig('US President Scatter Plot.png')
```

```python
# Using KMeans algorithm, we may standardize our data to identify our clusters

# Does it true that preprocessing.scale only accept numeric value?
x = preprocessing.scale(us_president[['height', 'age']])
# KMeans (x), where the x is for the cluster we want to classify
km = KMeans(4)
km.fit(x)
```

```python
plt.scatter(us_president['height'], us_president['age'], c=km.fit_predict(x), cmap='rainbow')
plt.xlabel('height')
plt.ylabel('age')
plt.title('KMeans US President Scatter Plot')
plt.savefig('KMeans US President Scatter Plot.png')
```

From the figure above, we have already obtained the four clusters of US President based on their age and their height. On the yellow cluster, it shows that president with the age 40 - 50 have the minimum height around 170-175 and the maximum height is around 185 - 190. Meanwhile the purple cluster, shows that most of the president with the age of 50s to 60s is shorter rather than most of all cluster. And the blue one shows presidents in their 50s to 60 are the tallest in their age. Lastly, the red cluster shows that elderly president are most of them is tall
