# Dummy variables

## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## Load the data

```python
raw_data = pd.read_csv('1.03. Dummies.csv')
```

```python
raw_data
```

## Predictor
Students with attendance more than 75% will be calculated as yes, otherwise will be calculated as no. As we know, that in normalization we need to transfrom the data to be categorical variable, thus we mapped attendance by using 1 and 0.

## Map the data

```python
data = raw_data.copy()
```

```python
data['Attendance'] = data['Attendance'].map({'Yes':1,'No': 0})
data
```

```python
data.describe()
```

As per the result shows above, we could interpret that 46% students has attended more than 75% of the class.

## Regression

```python
y = data['GPA']
x1 = data[['SAT','Attendance']]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

As you can see the P-Value of SAT and Attendance is significant, which means that these two variables is correlated with the target variable significantly.

Thus our model will be:

$
GPA = 0.6439 + 0.0014 * SAT + 0.2226 * Attendance
$

If the students categorized as 0, the model should looked like this:

$
GPA = 0.6439 + 0.0014 * SAT + 0.2226 * Attendance \\
GPA = 0.6439 + 0.0014 * SAT + 0.2226 * 0 \\
GPA = 0.6439 + 0.0014 * SAT
$

Otherwise, if the students categorized as 1, the model will be changed to be:

$
GPA = 0.6439 + 0.0014 * SAT + 0.2226 * Attendance \\
GPA = 0.6439 + 0.0014 * SAT + 0.2226 * 1 \\
GPA = 0.8665 + 0.0014 * SAT
$

## Plot the regression line(s) on the scatter plot

```python
plt.scatter(data['SAT'],y)
yhat_no = 0.6439 + 0.0014*data['SAT']
yhat_yes = 0.8665 + 0.0014*data['SAT']
fig = plt.plot(data['SAT'],yhat_no, lw=2, c='#006837')
fig = plt.plot(data['SAT'],yhat_yes, lw=2, c='#a50026')
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.legend(['Observation', 'Not Attended', 'Attended'],loc='lower right', fontsize=8)
plt.show()
```

## Plot the regression line(s) on the scatter plot and colour the data points

```python
plt.scatter(data['SAT'],y,c=data['Attendance'],cmap='RdYlGn_r')
yhat_no = 0.6439 + 0.0014*data['SAT']
yhat_yes = 0.8665 + 0.0014*data['SAT']
fig = plt.plot(data['SAT'],yhat_no, lw=2, c='#006837')
fig = plt.plot(data['SAT'],yhat_yes, lw=2, c='#a50026')
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.legend(['Observation', 'Not Attended', 'Attended'],loc='lower right', fontsize=8)
plt.show()
```
