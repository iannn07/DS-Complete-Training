# Simple linear regression

## Import the relevant libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

### Load the data

```python
data = pd.read_csv('Multiple linear regression.csv')
data.head()
```

```python
data.describe()
```

We want to predict the GPA based on their SAT Score and Rand 1, 2, 3. We knew that Rand variable isn't relevant but let's try to build a MLR with it.

## Define the dependent and the independent variables

```python
y = data['GPA']
x1 = data[['SAT', 'Rand 1,2,3']]
```

### EDA

```python
x = sm.add_constant(x1)
results = sm.OLS(y, x).fit()
results.summary()
```

Compare it with the old one, which is without Rand 1, 2, 3

```python
x2 = data['SAT']
```

```python
x_old = sm.add_constant(x2)
results_old = sm.OLS(y, x_old).fit()
results_old.summary()
```

See the difference of those R-Squared? It's indeed increased 0.001, but we've just lost the value since the Adjusted R-Squared is decreased. This means that we were penalized for adding an useless variable, which don't have a strong explanatory power. Other than that, Rand variable also insignificant
